```python
import causalgraphicalmodels as cgm
import networkx as nx
```

# Base Causal Graph

```python
g = cgm.CausalGraphicalModel(
    nodes=['Time', 'Popularity', 'Text', 'Domain', 'User'],
    edges=[
        ('Time', 'Popularity'),
        ('User', 'Time'),
        ('User', 'Text'),
        ('User', 'Domain'),
        ('Domain', 'Text'),
        ('Text', 'Popularity'),
    ]
)

g.draw()
```

```python
g.is_valid_backdoor_adjustment_set('Time', 'Popularity', {'Text', 'Domain', 'User'})
```

# Possible Hidden Covariates

```python
g = cgm.CausalGraphicalModel(
    nodes=['Time', 'Popularity', 'Text', 'Domain', 'User', 'H'],
    edges=[
        ('Time', 'Popularity'),
        ('User', 'Time'),
        ('User', 'Text'),
        ('User', 'Domain'),
        ('Domain', 'Text'),
        ('Text', 'Popularity'),
        # Hidden:
        ('H', 'Time'),
        ('H', 'Text')
    ]
)

g.is_valid_backdoor_adjustment_set('Time', 'Popularity', {'Text', 'Domain', 'User'})
```
